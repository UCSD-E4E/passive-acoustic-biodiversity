```python
def get_frame_number(uid):
    return int(uid.split("_")[0])
def get_window_number(uid):
    return int(uid.split("_")[1])
def get_start_pos(uid):
    return int(uid.split("_")[2])
def get_end_pos(uid):
    return int(uid.split("_")[3])
def get_filename(uid):
    return "".join(uid.split("_")[4:])
def get_id(uid):
    return "_".join(uid.split("_")[1:])
```

```python
the_uid = df.iloc[0]["uid"]
```

```python
get_frame_number(the_uid),get_window_number(the_uid),get_start_pos(the_uid),get_end_pos(the_uid),get_filename(the_uid), get_id(the_uid)
```

```python
frame_number = []
window_number = []
start_pos = []
end_pos = []
filename = []
nuid = []
for uid in df["uid"]:
    frame_number.append(get_frame_number(uid))
    window_number.append(get_window_number(uid))
    start_pos.append(get_start_pos(uid))
    end_pos.append(get_end_pos(uid))
    filename.append(get_filename(uid))
    nuid.append(get_id(uid))
```

```python
#do this a better way
#like macrolly
def find_dupes(df, unique_filenames, max_pos):
    dupes = []
    for i, f in enumerate(unique_filenames):
        for j in range(max_pos[i]):
            mini_df = df[df["filename"] == f]
            small_df = mini_df[mini_df["actual_frame"] == j]
            if len(small_df) > 1:
                dupes = small_df
    return dupes
```
