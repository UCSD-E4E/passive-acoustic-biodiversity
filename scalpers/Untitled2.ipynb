```python
import pandas as pd
import numpy as np
```

```python
cosmos_data = pd.read_csv("C:/Users/Siloux/Desktop/all_meta_data.csv")
```

```python
species = ["Blue-grey Tanager",
"Buff-throated Woodcreeper",
"Green-backed Trogon",
"Pale-breasted Thrush",
"Rufous-collared Sparrow",
"Scaly-breasted Woodpecker",
"Slate-throated Whitestart",
"Southern Nightingale-Wren",
"White-throated Toucan",
"Yellow-olive Flatbill"]
```

```python
cosmos_data = cosmos_data[cosmos_data["Country"].isin(["Brazil", "Bolivia", "Peru", "Ecuador", "Colombia", "Venezuela", "Guyana", "Suriname", "French Guiana"])]
cosmos_data = cosmos_data[cosmos_data["Common"].isin(species)]
cosmos_data
```

```python
cosmos_data["TIME"] = cosmos_data["Length"].apply(lambda x: int(x.split(":")[0]) * 60 + int(x.split(":")[1]))
```

```python
data_to_upload = pd.DataFrame(columns=cosmos_data.columns)
for species in np.unique(cosmos_data["Common"]):
    tmp_data_to_upload = pd.DataFrame(columns=cosmos_data.columns)
    tmp_df = cosmos_data[cosmos_data["Common"] == species]
    while (sum(tmp_data_to_upload["TIME"]) < 2000):
          tmp_df = tmp_df[~tmp_df["filename"].isin(tmp_data_to_upload["filename"])]
          file = tmp_df.sample(1)
          tmp_data_to_upload = tmp_data_to_upload.append(file)
    print(sum(tmp_data_to_upload["TIME"]))
    data_to_upload = data_to_upload.append(tmp_data_to_upload)
data_to_upload
```

```python
test = data_to_upload.groupby("filename").count()
test[test["Common"] > 1]
```
