```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading in the annotations created by the Torrey Pines State Reserve Volunteers
scripps_df0 = pd.read_csv("Scripps0.csv")
scripps_df1 = pd.read_csv("Scripps1.csv")
```

```python
manual_df = scripps_df0.append(scripps_df1)
manual_df.reset_index(drop=True,inplace=True)
manual_df
```

```python
# Removing the date from the column with username
for row in manual_df.index:
    full_string = manual_df["LAST MOD BY"][row]
    temp_list = full_string.split(" ")
    manual_df["LAST MOD BY"][row] = temp_list[0]
    manual_df["LAST MOD DATE"][row] = temp_list[1]
```

```python
# Removing undesirable annotations

# Removing all the annotations that were made by Jacob
#manual_df = manual_df[manual_df["LAST MOD BY"] != "JacobGlennAyers"]

# Removing all clips without birds and removing binary labels

manual_df = manual_df[manual_df["MANUAL ID"] != "1 Birds"] 
manual_df = manual_df[manual_df["MANUAL ID"] != "Birds"]
manual_df = manual_df[manual_df["MANUAL ID"] != "1 No Birds"]
manual_df = manual_df[manual_df["MANUAL ID"] != "1 Birds"]
```

```python
# Finding the clips that intersect the two datasets
automated_clip_list = automated_df["IN FILE"].to_list()
automated_clip_list = list(dict.fromkeys(automated_clip_list))
manual_clip_list = manual_df["IN FILE"].to_list()
manual_clip_list = list(dict.fromkeys(manual_clip_list))
print("Automated Clip Count: ",str(len(automated_clip_list)))
print("Manual Clip Count: ",str(len(manual_clip_list)))
```

```python
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```
