```python
import subprocess
import os
# subprocess.call(['python', 'somescript.py', somescript_arg1, somescript_val1,...]).
def getSeparateWaveForm(inputFileName, outputFileName):
    out_file = outputFileName + '.wav'
    '''
        use opso to resample to 22.1kHz for both inputfile and save in temp file
    '''
    audio_object_resample = Audio.from_file(inputFileName, sample_rate=22100)
    inputFileNameTemp = inputFileName.split('.wav')[0] + '_temp.wav'
    audio_object_resample.save(inputFileNameTemp)
    subprocess.call(['python', '../models/tools/process_wav.py', '--model_dir', '../bird_mixit_model_checkpoints/output_sources4',
                     '--checkpoint', '../bird_mixit_model_checkpoints/output_sources4/model.ckpt-3223090',
                     '--num_sources', '4',
                     '--input', inputFileNameTemp,
                     '--output', out_file
                    ])
    os.remove(inputFileNameTemp)
    out_files = [outputFileName+'_source0.wav', outputFileName+'_source1.wav', outputFileName+'_source2.wav', outputFileName+'_source3.wav']
    return out_files
```

```python
'''
python3 ../tools/process_wav.py \
--model_dir bird_mixit_model_checkpoints/output_sources4 \
--checkpoint bird_mixit_model_checkpoints/output_sources4/model.ckpt-3223090 \
--num_sources 4 \
--input <input name>.wav \
--output <output_name>.wav
'''
'''
../models/tools/process_wav.py
'''
df_xeno_canto_test = s1
for idx, rows in s1.iterrows():
    outputFileName = rows['Filename'].split('.wav')[0] + '_output'
    fileNames = getSeparateWaveForm(rows['Filename'], outputFileName)
    df_xeno_canto_test = df_xeno_canto_test.append({'Filename': fileNames[0], 'Bird(s)?': rows['Bird(s)?']}, ignore_index=True)
    df_xeno_canto_test = df_xeno_canto_test.append({'Filename': fileNames[1], 'Bird(s)?': rows['Bird(s)?']}, ignore_index=True)
    df_xeno_canto_test = df_xeno_canto_test.append({'Filename': fileNames[2], 'Bird(s)?': rows['Bird(s)?']}, ignore_index=True)
    df_xeno_canto_test = df_xeno_canto_test.append({'Filename': fileNames[3], 'Bird(s)?': rows['Bird(s)?']}, ignore_index=True)
```
