```python
def Move_Files_Custom(filename):
    file = filename.split("/")
    og_file_path = "C:/Users/Siloux/Desktop/E4E/passive-acoustic-biodiversity/OpenSoundScape/MultiClass Classifier/Mixed_Bird/"+ file[-1] 
    new_file_path = "C:/Users/Siloux/Desktop/E4E/passive-acoustic-biodiversity/OpenSoundScape/MultiClass Classifier/Cosmos_Data/Training_Xeno_Canto_2022/" + file[-1]
    if (not os.path.exists(og_file_path)):
        return "DELETE" 
    if(not os.path.exists(new_file_path)):
        print(new_file_path)
        shutil.copyfile(og_file_path, new_file_path)
        return new_file_path
    

Mixed["FILE_PATH"] = Mixed["FILE_PATH"].apply(Move_Files_Custom)
```

```python
Mixed["MANUAL ID"] = Mixed["MANUAL ID"].apply(lambda x: " ".join(x.split(" ")[:2]))
```
