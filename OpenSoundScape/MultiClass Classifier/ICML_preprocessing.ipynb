```python
import pandas as pd
import os
import shutil
```

```python


IMCL = pd.read_csv("C:/Users/Siloux/Desktop/E4E/passive-acoustic-biodiversity/OpenSoundScape/MultiClass Classifier/All/Copy of Stratified_Random_Sample_Bird_Labels.csv")
IMCL = IMCL[IMCL["Bird(s)?"] == "n"].sample(3000)
IMCL
```

```python
IMCL["FOLDER"] = IMCL["Folder"]
IMCL["IN FILE"] = IMCL["Clip"]
IMCL["MANUAL ID"] = IMCL["Bird(s)?"]
```

```python
IMCL["FILE_PATH"] = "/" +IMCL["FOLDER"] + "/" + IMCL["IN FILE"]
```

```python
def Move_Files_Custom(filename):
    file = filename.split("/")
    og_file_path = "C:/Users/Siloux/Desktop/E4E/passive-acoustic-biodiversity/OpenSoundScape/MultiClass Classifier/All/" + file[-2] + "/" + file[-1] 
    new_file_path = "C:/Users/Siloux/Desktop/E4E/passive-acoustic-biodiversity/OpenSoundScape/MultiClass Classifier/ICML/" + file[-2]+"_"+file[-1]
    if (not os.path.exists(og_file_path)):
        return "DELETE"
    if(not os.path.exists(new_file_path)):
        shutil.copyfile(og_file_path, new_file_path)
    return new_file_path

IMCL["FILE_PATH"] = IMCL["FILE_PATH"].apply(Move_Files_Custom)
```

```python
IMCL = IMCL[IMCL["FILE_PATH"] != "DELETE"]
IMCL
```

```python
no_bird = IMCL.reset_index(drop=True)
no_bird["FILENAME"] = no_bird["IN FILE"]+".wav"
no_bird["IN FILE"] = 'Training_Xeno_Canto_2022/' + no_bird["FOLDER"] + "_"+ no_bird["FILENAME"]
no_bird["OFFSET"] = float(0)
no_bird["DURATION"] = float(3)
no_bird["FOLDER"] = "C:/Users/Siloux/Desktop/E4E/passive-acoustic-biodiversity/OpenSoundScape/MultiClass Classifier/Cosmos_Data/"
no_bird["CHANNEL"] = 0
no_bird["CLIP LENGTH"] = float(3)
no_bird["MANUAL ID"] = "no bird"
no_bird["CONFIDENCE"] = 1
no_bird["SAMPLE RATE"] = 44100
```
